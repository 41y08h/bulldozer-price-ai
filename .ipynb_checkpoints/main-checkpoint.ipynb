```python
df = pd.read_csv("data/TrainAndValid.csv", low_memory=False, parse_dates=["saledate"])
df.head().T
```

```python
df.saledate.head()
```

```python
df_temp = df.copy()
df_temp = df.sort_values("saledate", ascending=True)
```

```python
df_temp.head().T
```

```python
df_temp["saleYear"] = df_temp.saledate.dt.year
df_temp["saleMonth"] = df_temp.saledate.dt.month
df_temp["saleDay"] = df_temp.saledate.dt.day
df_temp["saleDayOfWeek"] = df_temp.saledate.dt.day_of_week
df_temp["saleDayOfYear"] = df_temp.saledate.dt.day_of_year
```

```python
df_temp.drop("saledate", axis=1, inplace=True)
```

```python
df_temp
```

```python
df_temp.state.value_counts().plot.bar()
```

```python
fig, ax = plt.subplots()
ax.scatter(df.saledate[:1000], df["SalePrice"][:1000]);
```

```python
df_temp.saleYear.value_counts().plot.bar()
```

```python
# Transforming string data into categorical dtype

for label, content in df_temp.items():
    if(pd.api.types.is_string_dtype(content)):
        df_temp[label] = pd.Categorical(df_temp[label])        
```

```python
from sklearn.linear_model import SGDRegressor

model = SGDRegressor()
model.fit(df_temp.drop("SalePrice", axis=1), df_temp.SalePrice)
```
